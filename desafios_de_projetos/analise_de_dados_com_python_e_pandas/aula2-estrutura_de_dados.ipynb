# Tuplas

```python
# As tuplas usam parênteses como sintaxe.
tp = ("Banana", "Maçã", 10, 50)
```

```python
# Retornando o primeiro elemento
tp[0]
```

```python
# Diferente das listas, as tuplas são imutáveis. O que quer dizer que não podemos alterar os seus elementos.
tp[0] = "Laranja"
```

```python
tp.count("Maçã")
```

```python
tp[0:2]
```

# Dicionários

```python
# Para criar um dicionário utilizamos {}
# Dicionários trabalham com o conceito chave:valor
dc = {"Maçã": 20, "Banana": 10, "Laranja": 15, "Maçã": 20, "Uva": 5}
```

```python
dc
```

```python
# Acessando o valor de um dicionário através da chave.
dc["Maçã"]
```

```python
# Atualizando o valor de Maçã.
dc["Maçã"] = 25
dc
```

```python
# Retornando todas as chaves do dicionário.
dc.keys()
```

```python
# Retornando os valores do dicionário.
dc.values()
```

```python
# Verificando se já existe uma chave no dicioário e caso não exista, insira.
dc.setdefault("Limão", 22)
```
